In the next step, plot separate histograms for all seasons between 2000 and 2016. Ensure that the range of scores and categories in all histograms is consistent and spans the entire dataset from all seasons. Do you observe any specific trends over time?


```python
import pandas as pd
import matplotlib.pyplot as plt

def season_year(n: int):
    if n < 9:
        return f'200{n}-0{n+1}'
    elif n == 9:
        return '2009-10'
    return f'20{n}-{n+1:02d}'

data = pd.read_csv('/content/drive/My Drive/LaLiga_dataset.csv')
fig, axs = plt.subplots(4, 4, figsize=(10, 10))
season_data = [0]*16

for n in range(16):
    a = n // 4
    b = n % 4
    seasonn_data = data[data['season'] == season_year(n)]
    season_data[n] = seasonn_data['points'].tolist()
    #print(season_year(n) + '----for n=' + str(n))
    #print(season_data)
    min_range = min(season_data[n])
    max_range = max(season_data[n])
    axs[a, b].hist(season_data[n], bins=30, color='skyblue', edgecolor='black', range=(min_range, max_range))
    axs[a, b].set_title('season ' + season_year(n))
    axs[a,b].grid(True)


plt.tight_layout()
plt.show()
```

As we can see, the distribution of points before 2008 are like each other. consequently, the distribution of points in years after 2008 appears to be alike to some extent as well.

Now we want to use a numerical criterion to compare histograms

chi-square-distance1.jpg

Now let’s manually calculate this criterion between the two existing histograms and present the results in a 16x16 table. Explain your interpretation of the results

```python
# to do
import seaborn as sns


def chi_dist(a,b):
  if len(a)!=len(b):
    raise ValueError("Input lists must have the same length.")
  res = 0
  for i in range(len(a)):
    res += (.5) *((a[i] - b[i])**2 )/(a[i]+b[i])
  return res

cov_mat =  np.zeros((16, 16))
for n in range(256):
  #print(n)
  if n not in [0,3,8,15,24,35,48,63,80,99,120,143,168,195,224,255]:
    a = n//16
    b = n%16
    cov_mat[a][b] = chi_dist(season_data[a],season_data[b])

fig, axs = plt.subplots(1, 2, figsize=(20, 8))
axs[0].imshow(cov_mat, cmap='viridis', interpolation='none')
fig.colorbar(axs[0].images[0], ax=axs[0])
sns.heatmap(cov_mat, annot=True, fmt=".2f", cmap='viridis', ax=axs[1])

plt.show()
```

As we can see, the chi-square distances between the seasons from 2000 to 2008 are relatively small, indicating that these seasons are more similar to each other in terms of point distribution. In contrast, the seasons from 2009 to 2016 show larger chi-square distances, suggesting less similarity among them.

Therefore, we can deduce that the point distributions in the years 2000 to 2008 are more similar to each other compared to the years 2009 to 2016. This implies a noticeable shift in the distribution of points starting from the 2009 season.

Complete the functions:
`min_points(df_laliga,season)`
and
`max_points(df_laliga,season)`.

Note that it is possible for the maximum (or minimum) scores in a year to be equal for two teams. In this case, the team with the better (or worse) goal difference is considered as the output of the function



```python
def min_points(df_laliga, season):
    season_df = df_laliga[df_laliga['season'] == season]
    min_row = season_df.loc[season_df['points'].idxmin()]
    return {'club': min_row['club'], 'points': min_row['points']}
```

```python
def max_points(df_laliga, season):
    season_df = df_laliga[df_laliga['season'] == season]
    max_row = season_df.loc[season_df['points'].idxmax()]
    return {'club': max_row['club'], 'points': max_row['points']}
```

```python
import pandas as pd

pd.DataFrame.from_dict({'club': {
  '1970-71': 16,
  '1971-72': 18,
  '1972-73': 18,
  '1973-74': 18,
  '1974-75': 18}})

# to do
#df_laliga = pd.read_csv('/content/drive/My Drive/LaLiga_dataset.csv')
df_laliga = pd.read_csv('/home/parsa/Downloads/HW1/LaLiga_dataset.csv')
seasons = ['1970-71', '1971-72', '1972-73', '1973-74', '1974-75']
clubs_no = []

for season in seasons:
    x = df_laliga[df_laliga['season'] == season]
    clubs_no.append(len(x['club']))

my_dict = dict(zip(seasons, clubs_no))
club_count = pd.DataFrame.from_dict(my_dict, orient='index', columns=['Club_Count'])
club_count.index.name = 'Season'
club_count.reset_index(inplace=True)

club_count
```

Now we want to get information about the champion of different La Liga seasons. Therefore, we must first determine the champion of each season.

You can see the champion of the last 5 seasons of Dataset in the table below.

```python
pd.DataFrame.from_dict({'club': {'2012-13': 'Barcelona',
  '2013-14': 'Atletico de Madrid',
  '2014-15': 'Barcelona',
  '2015-16': 'Barcelona',
  '2016-17': 'Real Madrid'}})

#data = pd.read_csv('/content/drive/My Drive/LaLiga_dataset.csv')
data = pd.read_csv('/home/parsa/Downloads/HW1/LaLiga_dataset.csv')
seasons = ['2012-13','2013-14','2014-15','2015-16','2016-17']
champions = []

for season in seasons:
    season_df = data[data['season'] == season]
    max_row = data.loc[season_df['points'].idxmax()]
    champions.append(max_row['club'])

df_champions = pd.DataFrame({
    'Season': seasons,
    'Champion': champions
})

df_champions
```

```python
import zipfile
import os


def compress(file_names):
    print("File Paths:")
    print(file_names)

    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED

    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = [
    file for file in os.listdir()
    if (file.endswith('csv') or file.endswith('ipynb'))
    and file != 'LaLiga_dataset.csv'
]

compress(file_names)
```
