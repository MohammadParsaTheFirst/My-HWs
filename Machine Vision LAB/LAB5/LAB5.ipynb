```python

image1_gray = cv2.imread('image1_gray.jpg', cv2.IMREAD_GRAYSCALE)
image2_gray = cv2.imread('image2_gray.jpg', cv2.IMREAD_GRAYSCALE)

filter_sizes = [3, 5, 7]

average_filtered_images = []
median_filtered_images = []
gaussian_filtered_images = []

# Apply filters for each size
for size in filter_sizes:
    avg_filter = cv2.blur(image1_gray, (size, size))
    average_filtered_images.append(avg_filter)

    med_filter = cv2.medianBlur(image1_gray, size)
    median_filtered_images.append(med_filter)

    gauss_filter = cv2.GaussianBlur(image1_gray, (size, size), 0)
    gaussian_filtered_images.append(gauss_filter)

# Display results
plt.figure(figsize=(15, 10))

# Original Image
plt.subplot(4, 3, 2)
plt.imshow(image1_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Average Filter Results
for i in range(len(filter_sizes)):# 4 7 10 =3(i+1)+1   0 1 2
    plt.subplot(4, 3, 3*i + 4)
    plt.imshow(average_filtered_images[i], cmap='gray')
    plt.title(f'Average Filter {filter_sizes[i]}x{filter_sizes[i]}')
    plt.axis('off')

# Median Filter Results
for i in range(len(filter_sizes)):
    plt.subplot(4, 3, 3*i+5)
    plt.imshow(median_filtered_images[i], cmap='gray')
    plt.title(f'Median Filter {filter_sizes[i]}x{filter_sizes[i]}')
    plt.axis('off')

# Gaussian Filter Results
for i in range(len(filter_sizes)):
    plt.subplot(4, 3, 3*i+6)
    plt.imshow(gaussian_filtered_images[i], cmap='gray')
    plt.title(f'Gaussian Filter {filter_sizes[i]}x{filter_sizes[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def parsa1(img1, img2, threshold):
    if img1.shape != img2.shape:
        raise ValueError(f"Images must be the same size. Got {img1.shape} and {img2.shape}")
    
    diff_image = cv2.absdiff(img1, img2)
    _, bin_image = cv2.threshold(diff_image, threshold, 255, cv2.THRESH_BINARY)
    kernel = np.ones((9, 9), np.uint8)
    eroded_img = cv2.erode(bin_image, kernel, iterations=1)
    dilated_img = cv2.dilate(eroded_img, kernel, iterations=1)
    
    masked_img = cv2.bitwise_and(img1, img1, mask=dilated_img)
    
    return masked_img

def record_webcam(video_path, duration=5):
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        raise Exception("Could not open webcam")
    
    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = 20  # Frames per second, adjust as needed
    
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(video_path, fourcc, fps, (frame_width, frame_height))
    
    frames_recorded = 0
    max_frames = duration * fps
    
    while frames_recorded < max_frames:
        ret, frame = cap.read()
        if not ret:
            break
        out.write(frame)
        frames_recorded += 1
    
    # Release the capture
    cap.release()
    out.release()
    cv2.destroyAllWindows()

def process_video(video_path, background_path, output_path, threshold):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    # Read and resize the background image to match the video frame size
    background = cv2.imread(background_path)
    background_resized = cv2.resize(background, (frame_width, frame_height))
    background_gray = cv2.cvtColor(background_resized, cv2.COLOR_RGB2GRAY)
    print(f"Background shape: {background_gray.shape}")
    
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height), False)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Convert the frame to grayscale
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        #print(f"Frame shape: {frame_gray.shape}")
        
        # Apply the background removal function
        try:
            processed_frame = parsa1(frame_gray, background_gray, threshold)
        except ValueError as e:
            print(e)
            continue
        
        # Write the processed frame to the output video
        out.write(processed_frame)
    
    # Release everything when done
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# File paths
input_video_path = 'input.avi'
output_video_path = 'output.avi'
background_path = 'background.jpg'

# Step 1: Record from the webcam
record_webcam(input_video_path, duration=5)

# Step 2: Process the recorded video
process_video(input_video_path, background_path, output_video_path, threshold=60)
```
